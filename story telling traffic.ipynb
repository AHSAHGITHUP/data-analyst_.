```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("traffic.csv",sep=";")
```

```python
dataset.head()
```

```python
dataset["Slowness in traffic (%)"]=dataset["Slowness in traffic (%)"].str.replace(",",".").astype(float)
```

```python
dataset["Hour (Coded)"].value_counts().sort_values().head(2)
```

```python
days=["Monday","Tuesday","Wednesday","Thursday","Friday"]
dataset_per_day={}
for i,day in zip(range(0,135,27),days):
    dataset_per_day[day]=dataset.iloc[i:i+27]
    
```

```python
fig,(ax1,ax2,ax3,ax4,ax5)=plt.subplots(5,1,figsize=(10,12))
axis=ax1,ax2,ax3,ax4,ax5
for ax,day in zip(axis,days):
    ax.plot(dataset_per_day[day]["Hour (Coded)"],dataset_per_day[day]["Slowness in traffic (%)"])
plt.show()
```
