```python
def Match(year):
    Season=[]
    for f in dataset:
        season=f[-1]
        if f[-1]==year:
            Season.append(f)
    teams=[]
    for i in dataset:
        name=i[-1]
        if i[-1]==year:
            if i[2] not in teams:
                teams.append(i[2])
    Table=[]
    for i in teams:
        Table.append([i,0,0,0,0,0,0,0,0]) 
    
    
    for s in Season:
        home=s[2]
        away=s[3]
        hgoal=int(s[4])
        agoal=int(s[5])
        result=s[6]
        
        for t in Table:
            if result=="H":
                if home==t[0]:
                    t[1]+=1
                    t[2]+=1
                    t[5]+=hgoal
                    t[6]+=agoal
                    t[7]+=hgoal-agoal
                    t[8]+=3
                if away==t[0]:
                    t[1]+=1
                    t[3]+=1
                    t[5]+=agoal
                    t[6]+=hgoal
                    t[7]+=agoal-hgoal
                    t[8]+=0
                    
                    
            elif result=="A":
                if home==t[0]:
                    t[1]+=1
                    t[3]+=1
                    t[5]+=hgoal
                    t[6]+=agoal
                    t[7]+=hgoal-agoal
                    t[8]+=0
                if away==t[0]:
                    t[1]+=1
                    t[2]+=1
                    t[5]+=agoal
                    t[6]+=hgoal
                    t[7]+=agoal-hgoal
                    t[8]+=3
                    
                    
            elif result=="D":
                if home==t[0]:
                    t[1]+=1
                    t[4]+=1
                    t[5]+=hgoal
                    t[6]+=agoal
                    t[7]+=hgoal-agoal
                    t[8]+=1
                if away==t[0]:
                    t[1]+=1
                    t[4]+=1
                    t[5]+=agoal
                    t[6]+=hgoal
                    t[7]+=agoal-hgoal
                    t[8]+=1
    set=[]
    for r in range(1,21):
        set.append(r)
        
    import pandas as pd                
                
    Ebl=pd.DataFrame(Table)
    Ebl.columns=["Team","MP","W","L","D","GF","GA","GD","PTS"]
    Ebl=Ebl.sort_values(by="PTS",ascending=False).reset_index().drop("index",axis=1)
    Ebl.insert(0,"Rank",set)
    return Ebl
                    
            
                    
            
                    
            
        
        
        
    
```
